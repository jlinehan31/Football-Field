```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

%matplotlib inline
# %load_ext nb_black
```

```python
# Create a function to draw field parameters
def draw_field(ax):
    # Set field outline
    plt.plot([0, 0], [0, 53.33], c="white", lw=2)
    plt.plot([0, 120], [53.33, 53.33], c="white", lw=2)
    plt.plot([120, 120], [53.33, 0], c="white", lw=2)
    plt.plot([120, 0], [0, 0], c="white", lw=2)

    # Set end zones
    plt.plot([10, 10], [0, 53.33], c="white", lw=2)
    plt.plot([110, 110], [0, 53.33], c="white", lw=2)

    # Set 10-yard lines
    yard_line = 15
    for i in range(19):
        plt.plot([yard_line, yard_line], [0, 53], c="white")
        yard_line += 5

    # Set emphasis on 20-yard lines
    plt.plot([29.8, 29.8], [0.3, 53], c="blue")
    plt.plot([30.3, 30.3], [0.3, 53], c="blue")
    plt.plot([89.8, 89.8], [0.3, 53], c="blue")
    plt.plot([90.3, 90.3], [0.3, 53], c="blue")

    # Set hash marks
    x_pos = 11
    for i in range(100):
        plt.plot([x_pos, x_pos], [19.5, 20.5], c="white")
        x_pos += 1

    x_pos = 11
    for i in range(100):
        plt.plot([x_pos, x_pos], [32.83, 33.83], c="white")
        x_pos += 1

    x_pos = 11
    for i in range(100):
        plt.plot([x_pos, x_pos], [0, 1], c="white")
        x_pos += 1

    x_pos = 11
    for i in range(100):
        plt.plot([x_pos, x_pos], [53.33, 52.33], c="white")
        x_pos += 1

    # Set PAT snap mark
    plt.plot([13, 13], [26.34, 27], c="white")
    plt.plot([107, 107], [26.34, 27], c="white")

    # Set kickoff 'x'
    plt.plot([44.6, 45.4], [26.34, 27], c="white")
    plt.plot([44.6, 45.4], [27, 26.34], c="white")
    plt.plot([74.6, 75.4], [26.34, 27], c="white")
    plt.plot([74.6, 75.4], [27, 26.34], c="white")

    # Add goalposts
    plt.plot([0.3, 0.3], [23.47, 29.7], c="yellow", lw=3)
    plt.plot([119.6, 119.6], [23.47, 29.7], c="yellow", lw=3)

    # Add padding to surround the field
    plt.plot([-0.5, -0.5], [0, 53.33], c="white", lw=10)
    plt.plot([120.5, 120.5], [0, 53.33], c="white", lw=10)
    plt.plot([-0.5, 120.5], [54, 54], c="white", lw=10)
    plt.plot([-0.5, 120.5], [-0.72, -0.72], c="white", lw=10)

    # Add coaches box (padding)
    plt.plot([36.25, 83.75], [56, 56], c="white", lw=25)
    plt.plot([36.25, 83.75], [-2.5, -2.5], c="white", lw=25)

    # Add yardage markers
    yard_line = 17.5
    yard_number = 10
    for i in range(9):
        plt.text(
            yard_line, 7.5, yard_number, fontsize=36, fontweight="heavy", color="white"
        )
        if i > 3:
            yard_number -= 10
        else:
            yard_number += 10
        yard_line += 10

    yard_line = 16.5
    yard_number = 10
    for i in range(9):
        plt.text(
            yard_line,
            43.83,
            yard_number,
            fontsize=36,
            fontweight="heavy",
            color="white",
            rotation=180,
        )
        if i > 3:
            yard_number -= 10
        else:
            yard_number += 10
        yard_line += 10

    # Add end zone abbreviations
    plt.text(
        0.5,
        18.5,
        "BYU",
        fontsize=90,
        fontweight="extra bold",
        color="white",
        rotation=90,
    )
    plt.text(
        109.7,
        3.5,
        "COUGARS",
        fontsize=76,
        fontweight="extra bold",
        color="white",
        rotation=270,
    )

    # Set background color
    ax.patch.set_facecolor("#348C31")

    # Set center BYU logo
    byu_logo = mpimg.imread("byu_logo_oval.png")
    image_box = OffsetImage(byu_logo, zoom=0.15)
    ab = AnnotationBbox(image_box, (60, 26.5), frameon=False)
    ax.add_artist(ab)
```

```python
# Set up field viz
fig = plt.figure()
fig.set_size_inches(22.5, 10)
ax = fig.add_subplot(1, 1, 1)
plt.xticks([])
plt.yticks([])
plt.tick_params(bottom=False)
plt.tick_params(left=False)
draw_field(ax)

plt.show()
```
